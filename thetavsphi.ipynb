In the Bloch sphere, the polar angle and azimuthal angle are two fundamental parameters that uniquely define any quantum state position on the sphere's surface. Here are the key differences:

Polar Angle (θ)
Definition: The angle measured from the positive Z-axis down to the Bloch vector

Range: 0 to π (0° to 180°)

Physical Meaning:

Determines the inclination of the quantum state with respect to the Z-axis

Controls the amplitude ratio between |0⟩ and |1⟩ components

When θ = 0: Pure |0⟩ state (North pole)

When θ = π: Pure |1⟩ state (South pole)

When θ = π/2: Equal superposition states on the equator

Mathematical Role: Appears as cos(θ/2) and sin(θ/2) in the quantum state equation

Azimuthal Angle (φ)
Definition: The angle measured from the positive X-axis, rotating counterclockwise around the Z-axis

Range: 0 to 2π (0° to 360°)

Physical Meaning:

Determines the phase relationship between |0⟩ and |1⟩ components

Controls which type of superposition state you have on the equator

When φ = 0: |+⟩ state (positive X-axis)

When φ = π: |-⟩ state (negative X-axis)

When φ = π/2: |+i⟩ state (positive Y-axis)

When φ = 3π/2: |-i⟩ state (negative Y-axis)

Mathematical Role: Appears as e^(iφ) in the quantum state equation
