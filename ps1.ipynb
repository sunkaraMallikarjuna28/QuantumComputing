Concept 2: Density Matrix Construction and Pauli Matrix Decomposition
Based on your understanding of the first concept, let's move to the next fundamental concept from the transcript: how quantum states are represented using density matrices and decomposed using Pauli matrices.

Mathematical Foundation from the Transcript
The transcript shows that any quantum state can be written as:
**ρ = (1/2)(I + r⃗·σ⃗)**

Where:

ρ is the density matrix (2×2 matrix representing the quantum state)

I is the 2×2 identity matrix

r⃗ is the unit vector on the Bloch sphere with components (r_x, r_y, r_z)

σ⃗ is the vector of Pauli matrices (σ_x, σ_y, σ_z)

```python
# Concept 2: Density Matrix Construction and Pauli Decomposition
import numpy as np
from braket.circuits import Circuit
from braket.devices import LocalSimulator

# Initialize simulator
device = LocalSimulator("braket_sv")

# Define the Pauli matrices as mentioned in the transcript
def define_pauli_matrices():
    """
    Define the three Pauli matrices σ_x, σ_y, σ_z and identity matrix I
    These form the basis for decomposing any 2×2 Hermitian matrix
    """
    
    # Identity matrix I - leaves quantum states unchanged
    I = np.array([[1, 0],    # |0⟩ stays |0⟩
                  [0, 1]])   # |1⟩ stays |1⟩
    
    # Pauli-X matrix σ_x - bit flip operation (swaps |0⟩ ↔ |1⟩)
    sigma_x = np.array([[0, 1],    # |0⟩ → |1⟩
                        [1, 0]])   # |1⟩ → |0⟩
    
    # Pauli-Y matrix σ_y - bit flip + phase flip
    sigma_y = np.array([[0, -1j],  # |0⟩ → i|1⟩
                        [1j, 0]])  # |1⟩ → -i|0⟩
    
    # Pauli-Z matrix σ_z - phase flip (adds minus sign to |1⟩)
    sigma_z = np.array([[1, 0],    # |0⟩ → |0⟩
                        [0, -1]])  # |1⟩ → -|1⟩
    
    print("=== PAULI MATRICES DEFINED ===")
    print("Identity Matrix (I):")
    print(I)
    print("\nPauli-X Matrix (σ_x):")
    print(sigma_x)
    print("\nPauli-Y Matrix (σ_y):")
    print(sigma_y)
    print("\nPauli-Z Matrix (σ_z):")
    print(sigma_z)
    
    return I, sigma_x, sigma_y, sigma_z

def construct_density_matrix_from_statevector(statevector):
    """
    Construct density matrix ρ = |ψ⟩⟨ψ| from a quantum state vector
    
    This implements the mathematical operation described in the transcript:
    - Take the state vector |ψ⟩ = [α, β]
    - Create the bra ⟨ψ| by taking conjugate transpose
    - Multiply to get the density matrix
    
    Parameters:
    - statevector: complex array [α, β] representing quantum state
    
    Returns:
    - rho: 2×2 density matrix
    """
    
    # Convert statevector to column vector (ket |ψ⟩)
    # The transcript shows this as a column vector with two entries
    psi_ket = statevector.reshape(-1, 1)  # Shape: (2, 1)
    
    # Create bra ⟨ψ| by taking conjugate transpose
    # The transcript mentions "conjugate transpose" - this flips complex signs
    psi_bra = statevector.conj().reshape(1, -1)  # Shape: (1, 2)
    
    # Perform matrix multiplication |ψ⟩⟨ψ| to get density matrix
    # This gives us the 2×2 density matrix as shown in the transcript
    rho = psi_ket @ psi_bra  # Matrix multiplication: (2,1) × (1,2) = (2,2)
    
    print(f"State vector |ψ⟩: {statevector}")
    print(f"Bra ⟨ψ|: {psi_bra[0]}")
    print("Density matrix ρ = |ψ⟩⟨ψ|:")
    print(rho)
    
    return rho

def decompose_in_pauli_basis(rho, I, sigma_x, sigma_y, sigma_z):
    """
    Decompose density matrix in Pauli basis: ρ = (1/2)(I + r⃗·σ⃗)
    
    This implements the key formula from the transcript that shows how
    any quantum state can be written as a combination of Pauli matrices
    
    Parameters:
    - rho: 2×2 density matrix
    - I, sigma_x, sigma_y, sigma_z: Pauli matrices
    
    Returns:
    - coefficients: [c_I, c_x, c_y, c_z] for the decomposition
    """
    
    # Extract coefficients using trace formula: c_i = Tr(ρ σ_i)
    # The transcript shows that we can extract Bloch vector components this way
    
    # Coefficient of identity matrix (should be 1/2 for normalized states)
    c_I = np.trace(rho @ I) / 2
    
    # X-component of Bloch vector: r_x = Tr(ρ σ_x)
    c_x = np.trace(rho @ sigma_x)
    
    # Y-component of Bloch vector: r_y = Tr(ρ σ_y)  
    c_y = np.trace(rho @ sigma_y)
    
    # Z-component of Bloch vector: r_z = Tr(ρ σ_z)
    c_z = np.trace(rho @ sigma_z)
    
    print("\n=== PAULI DECOMPOSITION ===")
    print(f"ρ = {c_I:.4f}×I + {c_x:.4f}×σ_x + {c_y:.4f}×σ_y + {c_z:.4f}×σ_z")
    print(f"Bloch vector r⃗ = ({c_x:.4f}, {c_y:.4f}, {c_z:.4f})")
    
    # Verify the decomposition by reconstructing the density matrix
    rho_reconstructed = c_I * I + c_x * sigma_x + c_y * sigma_y + c_z * sigma_z
    reconstruction_error = np.max(np.abs(rho - rho_reconstructed))
    print(f"Reconstruction error: {reconstruction_error:.10f}")
    
    return [c_I, c_x, c_y, c_z]

def create_quantum_circuit_and_analyze(theta, phi, description):
    """
    Create quantum circuit, extract statevector, and perform complete analysis
    
    This connects the Amazon Braket implementation with the mathematical theory
    """
    
    print(f"\n{'='*60}")
    print(f"ANALYZING: {description}")
    print(f"Parameters: θ={theta:.3f}, φ={phi:.3f}")
    print(f"{'='*60}")
    
    # Create quantum circuit using rotation gates
    circuit = Circuit()
    circuit.ry(0, theta)  # Y-rotation for polar angle θ
    if abs(phi) > 1e-10:  # Only add Z-rotation if φ is significant
        circuit.rz(0, phi)  # Z-rotation for azimuthal angle φ
    
    # Add result type for exact statevector extraction
    circuit.state_vector()
    
    # Execute circuit and get statevector
    result = device.run(circuit, shots=0).result()
    statevector = np.array(result.values[0])
    
    # Construct density matrix from statevector
    rho = construct_density_matrix_from_statevector(statevector)
    
    # Decompose in Pauli basis
    coefficients = decompose_in_pauli_basis(rho, I, sigma_x, sigma_y, sigma_z)
    
    # Verify density matrix properties
    trace_rho = np.trace(rho)
    is_hermitian = np.allclose(rho, rho.conj().T)
    eigenvals = np.linalg.eigvals(rho)
    
    print(f"\nDensity Matrix Properties:")
    print(f"  Trace(ρ) = {trace_rho:.6f} (should be 1.0)")
    print(f"  Hermitian: {is_hermitian}")
    print(f"  Eigenvalues: {eigenvals}")
    print(f"  Pure state check Tr(ρ²): {np.trace(rho @ rho):.6f} (should be 1.0)")
    
    return statevector, rho, coefficients

# Execute the analysis
print("=== DENSITY MATRIX CONSTRUCTION AND PAULI DECOMPOSITION ===")

# Define Pauli matrices
I, sigma_x, sigma_y, sigma_z = define_pauli_matrices()

# Test cases from the transcript context
test_cases = [
    (0, 0, "Ground State |0⟩"),           # North pole of Bloch sphere
    (np.pi, 0, "Excited State |1⟩"),      # South pole of Bloch sphere  
    (np.pi/2, 0, "Superposition |+⟩"),    # +X eigenstate
    (np.pi/2, np.pi/2, "Complex State |+i⟩")  # +Y eigenstate
]

# Analyze each test case
analysis_results = []
for theta, phi, description in test_cases:
    statevector, rho, coeffs = create_quantum_circuit_and_analyze(theta, phi, description)
    analysis_results.append((description, statevector, rho, coeffs))
```

```python
# Concept 4: Hilbert Space to Real Space Mapping and Eigenstate Relationships
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from braket.circuits import Circuit
from braket.devices import LocalSimulator

# Initialize simulator
device = LocalSimulator("braket_sv")

def demonstrate_hilbert_to_real_mapping():
    """
    Demonstrate the fundamental mapping between Hilbert space and real space
    as described in the SRT transcript
    """
    
    print("=== HILBERT SPACE TO REAL SPACE MAPPING ===")
    print("From transcript: 'PSI belongs to Hilbert space...equivalent to two-dimensional complex space'")
    print("Mapping: 2D Complex Hilbert Space ↔ 3D Real Space (Bloch Sphere)")
    
    # Define test points covering the entire parameter space
    test_mappings = [
        # Special points from transcript
        (0, 0, "North Pole |0⟩", "Z eigenstate +1"),
        (np.pi, 0, "South Pole |1⟩", "Z eigenstate -1"),
        (np.pi/2, 0, "Positive X-axis |+⟩", "X eigenstate +1"),
        (np.pi/2, np.pi, "Negative X-axis |-⟩", "X eigenstate -1"),
        (np.pi/2, np.pi/2, "Positive Y-axis |+Y⟩", "Y eigenstate +1"),
        (np.pi/2, 3*np.pi/2, "Negative Y-axis |-Y⟩", "Y eigenstate -1"),
        
        # General points to show complete mapping
        (np.pi/4, np.pi/4, "General Point 1", "Mixed state"),
        (3*np.pi/4, 3*np.pi/4, "General Point 2", "Mixed state"),
        (np.pi/3, np.pi/6, "General Point 3", "Mixed state"),
        (2*np.pi/3, 5*np.pi/6, "General Point 4", "Mixed state")
    ]
    
    print("\nMapping Analysis:")
    print("Spherical (θ, φ) → Hilbert Space [α, β] → Real Space (r_x, r_y, r_z)")
    print("-" * 80)
    
    mapping_results = []
    
    for theta, phi, description, eigenstate_type in test_mappings:
        # Step 1: Spherical coordinates → Hilbert space representation
        # From transcript: quantum state with parameters theta and phi
        alpha = np.cos(theta/2)                    # Real amplitude for |0⟩
        beta = np.exp(1j * phi) * np.sin(theta/2)  # Complex amplitude for |1⟩
        hilbert_vector = np.array([alpha, beta])
        
        # Step 2: Create quantum circuit to verify
        circuit = Circuit()
        if theta == 0 and phi == 0:
            circuit.i(0)  # Identity for |0⟩ state
        else:
            circuit.ry(0, theta)  # Y-rotation for theta
            if abs(phi) > 1e-10:
                circuit.rz(0, phi)  # Z-rotation for phi
        circuit.state_vector()
        
        # Get quantum statevector
        result = device.run(circuit, shots=0).result()
        quantum_statevector = np.array(result.values[0])
        
        # Step 3: Hilbert space → Real space (Bloch coordinates)
        # From transcript: "unit vector R which belongs to three-dimensional real space"
        r_x = 2 * np.real(np.conj(alpha) * beta)  # X-component
        r_y = 2 * np.imag(np.conj(alpha) * beta)  # Y-component
        r_z = np.abs(alpha)**2 - np.abs(beta)**2  # Z-component
        
        # Verify using geometric formulas from transcript
        r_x_geometric = np.sin(theta) * np.cos(phi)  # "sine Theta Cos Phi"
        r_y_geometric = np.sin(theta) * np.sin(phi)  # "sine Theta sine Phi"
        r_z_geometric = np.cos(theta)                # "Cos Theta"
        
        # Calculate mapping errors
        hilbert_error = np.linalg.norm(hilbert_vector - quantum_statevector)
        geometric_error = np.sqrt((r_x - r_x_geometric)**2 + 
                                (r_y - r_y_geometric)**2 + 
                                (r_z - r_z_geometric)**2)
        
        print(f"{description:<20} θ={theta:.3f} φ={phi:.3f}")
        print(f"  Hilbert: [{alpha:.4f}, {beta:.4f}] → Real: ({r_x:.4f}, {r_y:.4f}, {r_z:.4f})")
        print(f"  Geometric verification: ({r_x_geometric:.4f}, {r_y_geometric:.4f}, {r_z_geometric:.4f})")
        print(f"  Errors: Hilbert={hilbert_error:.2e}, Geometric={geometric_error:.2e}")
        print(f"  Type: {eigenstate_type}")
        
        mapping_results.append({
            'description': description,
            'theta': theta, 'phi': phi,
            'hilbert': hilbert_vector,
            'real': (r_x, r_y, r_z),
            'quantum': quantum_statevector,
            'type': eigenstate_type
        })
        print()
    
    return mapping_results

def verify_eigenstate_correspondence(mapping_results):
    """
    Verify the eigenstate relationships described in the transcript
    """
    
    print("=== EIGENSTATE CORRESPONDENCE VERIFICATION ===")
    print("From transcript: 'eigenvalue of X&Y are also plus one and -1'")
    
    # Define Pauli matrices
    sigma_x = np.array([[0, 1], [1, 0]])    # X Pauli matrix
    sigma_y = np.array([[0, -1j], [1j, 0]]) # Y Pauli matrix
    sigma_z = np.array([[1, 0], [0, -1]])   # Z Pauli matrix
    
    # Map eigenstate types to their corresponding operators and expected eigenvalues
    eigenstate_tests = [
        ("Z eigenstate +1", sigma_z, +1),
        ("Z eigenstate -1", sigma_z, -1),
        ("X eigenstate +1", sigma_x, +1),
        ("X eigenstate -1", sigma_x, -1),
        ("Y eigenstate +1", sigma_y, +1),
        ("Y eigenstate -1", sigma_y, -1)
    ]
    
    print("Verifying eigenstate equations: σ|ψ⟩ = λ|ψ⟩")
    print("-" * 60)
    
    for result in mapping_results:
        eigenstate_type = result['type']
        quantum_state = result['quantum']
        description = result['description']
        
        # Find matching eigenstate test
        for test_type, pauli_op, expected_eigenvalue in eigenstate_tests:
            if eigenstate_type == test_type:
                # Test eigenvalue equation
                sigma_psi = pauli_op @ quantum_state
                lambda_psi = expected_eigenvalue * quantum_state
                eigenvalue_error = np.linalg.norm(sigma_psi - lambda_psi)
                
                # Calculate actual eigenvalue
                actual_eigenvalue = np.conj(quantum_state) @ pauli_op @ quantum_state
                
                print(f"{description:<20}: σ|ψ⟩ = λ|ψ⟩")
                print(f"  Expected λ: {expected_eigenvalue:+2.0f}")
                print(f"  Actual λ:   {actual_eigenvalue.real:+6.4f}")
                print(f"  Error:      {eigenvalue_error:.2e}")
                
                # Verify Bloch sphere position
                r_x, r_y, r_z = result['real']
                if test_type == "Z eigenstate +1":
                    expected_pos = "North pole (0, 0, +1)"
                elif test_type == "Z eigenstate -1":
                    expected_pos = "South pole (0, 0, -1)"
                elif test_type == "X eigenstate +1":
                    expected_pos = "Positive X-axis (+1, 0, 0)"
                elif test_type == "X eigenstate -1":
                    expected_pos = "Negative X-axis (-1, 0, 0)"
                elif test_type == "Y eigenstate +1":
                    expected_pos = "Positive Y-axis (0, +1, 0)"
                elif test_type == "Y eigenstate -1":
                    expected_pos = "Negative Y-axis (0, -1, 0)"
                
                print(f"  Bloch position: ({r_x:.4f}, {r_y:.4f}, {r_z:.4f})")
                print(f"  Expected: {expected_pos}")
                print()
                break

def demonstrate_parameter_space_coverage():
    """
    Demonstrate how two real parameters (θ, φ) completely parameterize 
    the quantum state space as described in the transcript
    """
    
    print("=== PARAMETER SPACE COVERAGE ===")
    print("From transcript: 'you parameterize unit sphere by using these two Theta and Phi'")
    
    # Create a grid of (θ, φ) values to show complete coverage
    theta_values = np.linspace(0, np.pi, 8)      # Polar angle: 0 to π
    phi_values = np.linspace(0, 2*np.pi, 8)      # Azimuthal angle: 0 to 2π
    
    print(f"Testing {len(theta_values)} × {len(phi_values)} = {len(theta_values)*len(phi_values)} parameter combinations")
    print("Verifying: (θ, φ) → |ψ⟩ → (r_x, r_y, r_z) → (θ', φ') ≈ (θ, φ)")
    
    max_error_theta = 0
    max_error_phi = 0
    total_tests = 0
    
    for i, theta in enumerate(theta_values):
        for j, phi in enumerate(phi_values):
            # Forward mapping: (θ, φ) → |ψ⟩ → (r_x, r_y, r_z)
            alpha = np.cos(theta/2)
            beta = np.exp(1j * phi) * np.sin(theta/2)
            
            r_x = 2 * np.real(np.conj(alpha) * beta)
            r_y = 2 * np.imag(np.conj(alpha) * beta)
            r_z = np.abs(alpha)**2 - np.abs(beta)**2
            
            # Backward mapping: (r_x, r_y, r_z) → (θ', φ')
            magnitude = np.sqrt(r_x**2 + r_y**2 + r_z**2)
            
            if magnitude > 1e-10:
                theta_recovered = np.arccos(r_z / magnitude)
            else:
                theta_recovered = 0
            
            phi_recovered = np.arctan2(r_y, r_x)
            if phi_recovered < 0:
                phi_recovered += 2*np.pi
            
            # Calculate errors (handle periodic boundary for φ)
            theta_error = abs(theta - theta_recovered)
            phi_error = min(abs(phi - phi_recovered), 
                          abs(phi - phi_recovered + 2*np.pi),
                          abs(phi - phi_recovered - 2*np.pi))
            
            max_error_theta = max(max_error_theta, theta_error)
            max_error_phi = max(max_error_phi, phi_error)
            total_tests += 1
            
            # Print details for corner cases
            if (i == 0 or i == len(theta_values)-1) and (j == 0 or j == len(phi_values)-1):
                print(f"  θ={theta:.3f}→{theta_recovered:.3f} (Δ={theta_error:.2e}), " +
                      f"φ={phi:.3f}→{phi_recovered:.3f} (Δ={phi_error:.2e})")
    
    print(f"\nParameter Space Coverage Results:")
    print(f"  Total tests: {total_tests}")
    print(f"  Maximum θ error: {max_error_theta:.2e}")
    print(f"  Maximum φ error: {max_error_phi:.2e}")
    print(f"  ✓ Bidirectional mapping verified within numerical precision")

def create_hilbert_real_space_visualization(mapping_results):
    """
    Create side-by-side visualization of Hilbert space and real space representations
    """
    
    print("\n=== CREATING HILBERT ↔ REAL SPACE VISUALIZATION ===")
    
    fig = plt.figure(figsize=(16, 8))
    
    # Left plot: Hilbert space (complex plane representation)
    ax1 = fig.add_subplot(121)
    
    for result in mapping_results:
        hilbert_vec = result['hilbert']
        alpha, beta = hilbert_vec[0], hilbert_vec[1]
        description = result['description']
        
        # Plot alpha and beta as points in complex plane
        ax1.scatter(np.real(alpha), np.imag(alpha), s=100, alpha=0.7, 
                   label=f'{description[:8]} α')
        ax1.scatter(np.real(beta), np.imag(beta), s=100, alpha=0.7, 
                   marker='x', label=f'{description[:8]} β')
        
        # Add annotations
        ax1.annotate(f'{description[:8]}α', 
                    (np.real(alpha), np.imag(alpha)), 
                    xytext=(5, 5), textcoords='offset points', fontsize=8)
    
    ax1.set_xlabel('Real Part', fontsize=12)
    ax1.set_ylabel('Imaginary Part', fontsize=12)
    ax1.set_title('Hilbert Space\n(2D Complex Amplitudes)', fontsize=14, fontweight='bold')
    ax1.grid(True, alpha=0.3)
    ax1.set_aspect('equal')
    
    # Draw unit circle for normalization constraint
    circle = plt.Circle((0, 0), 1, fill=False, color='gray', alpha=0.5, linestyle='--')
    ax1.add_patch(circle)
    ax1.text(0.7, 0.7, '|α|² + |β|² = 1', fontsize=10, 
             bbox=dict(boxstyle="round,pad=0.3", facecolor="white", alpha=0.8))
    
    # Right plot: Real space (Bloch sphere)
    ax2 = fig.add_subplot(122, projection='3d')
    
    # Draw Bloch sphere wireframe
    u = np.linspace(0, 2 * np.pi, 20)
    v = np.linspace(0, np.pi, 20)
    sphere_x = np.outer(np.cos(u), np.sin(v))
    sphere_y = np.outer(np.sin(u), np.sin(v))
    sphere_z = np.outer(np.ones(np.size(u)), np.cos(v))
    ax2.plot_wireframe(sphere_x, sphere_y, sphere_z, alpha=0.1, color='lightgray')
    
    # Draw coordinate axes
    axis_length = 1.3
    ax2.plot([0, axis_length], [0, 0], [0, 0], 'k-', linewidth=2)  # X-axis
    ax2.plot([0, 0], [0, axis_length], [0, 0], 'k-', linewidth=2)  # Y-axis
    ax2.plot([0, 0], [0, 0], [0, axis_length], 'k-', linewidth=2)  # Z-axis
    
    # Label axes and special points from transcript
    ax2.text(axis_length, 0, 0, 'X', fontsize=12, fontweight='bold')
    ax2.text(0, axis_length, 0, 'Y', fontsize=12, fontweight='bold')
    ax2.text(0, 0, axis_length, 'Z', fontsize=12, fontweight='bold')
    ax2.text(0, 0, 1.1, '|0⟩', fontsize=12, ha='center', fontweight='bold')
    ax2.text(0, 0, -1.1, '|1⟩', fontsize=12, ha='center', fontweight='bold')
    
    # Plot quantum states on Bloch sphere
    colors = plt.cm.tab10(np.linspace(0, 1, len(mapping_results)))
    
    for i, result in enumerate(mapping_results):
        r_x, r_y, r_z = result['real']
        description = result['description']
        
        # Plot point and vector
        ax2.scatter(r_x, r_y, r_z, c=[colors[i]], s=100, alpha=0.8, edgecolors='black')
        ax2.plot([0, r_x], [0, r_y], [0, r_z], color=colors[i], linewidth=2, alpha=0.7)
        
        # Add label
        offset = 0.15
        ax2.text(r_x + offset, r_y + offset, r_z + offset, description[:8], 
                fontsize=10, color=colors[i], fontweight='bold')
    
    ax2.set_xlim([-1.4, 1.4])
    ax2.set_ylim([-1.4, 1.4])
    ax2.set_zlim([-1.4, 1.4])
    ax2.set_xlabel('X (Real)', fontsize=12)
    ax2.set_ylabel('Y (Real)', fontsize=12)
    ax2.set_zlabel('Z (Real)', fontsize=12)
    ax2.set_title('Real Space\n(3D Bloch Sphere)', fontsize=14, fontweight='bold')
    ax2.view_init(elev=20, azim=45)
    
    plt.tight_layout()
    plt.show()

# Execute the complete analysis
def main():
    """Execute complete Hilbert space to real space mapping analysis"""
    
    print("=== CONCEPT 4: HILBERT SPACE TO REAL SPACE MAPPING ===")
    print("Based on SRT transcript analysis of quantum state parameterization\n")
    
    # Demonstrate the fundamental mapping
    mapping_results = demonstrate_hilbert_to_real_mapping()
    
    # Verify eigenstate correspondence
    verify_eigenstate_correspondence(mapping_results)
    
    # Demonstrate complete parameter space coverage
    demonstrate_parameter_space_coverage()
    
    # Create visualization
    create_hilbert_real_space_visualization(mapping_results)
    
    print("\n=== KEY INSIGHTS FROM CONCEPT 4 ===")
    print("✓ 2D Complex Hilbert Space ↔ 3D Real Space mapping verified")
    print("✓ Two real parameters (θ, φ) completely parameterize quantum states")
    print("✓ Pauli eigenstates correspond to specific sphere points")
    print("✓ Bidirectional mapping preserves all quantum information")
    print("✓ Geometric and algebraic formulations are equivalent")

# Run the analysis
main()
```

```python
# Concept 5: Quantum Superposition and Measurement Probability Relationships
import numpy as np
import matplotlib.pyplot as plt
from braket.circuits import Circuit
from braket.devices import LocalSimulator

# Initialize simulator
device = LocalSimulator("braket_sv")

def analyze_superposition_probabilities():
    """
    Analyze how quantum superposition creates measurement probabilities
    as described in the SRT transcript
    """
    
    print("=== QUANTUM SUPERPOSITION AND MEASUREMENT PROBABILITIES ===")
    print("From transcript: 'Cos Theta by two square here, sine Theta by two square here'")
    print("Demonstrating: |ψ⟩ = cos(θ/2)|0⟩ + e^(iφ)sin(θ/2)|1⟩")
    
    # Test various superposition states from the transcript
    superposition_tests = [
        # Pure states (no superposition)
        (0, 0, "|0⟩ Pure State", "100% |0⟩, 0% |1⟩"),
        (np.pi, 0, "|1⟩ Pure State", "0% |0⟩, 100% |1⟩"),
        
        # Equal superposition states from transcript
        (np.pi/2, 0, "|+⟩ Equal Superposition", "50% |0⟩, 50% |1⟩"),
        (np.pi/2, np.pi, "|-⟩ Equal Superposition", "50% |0⟩, 50% |1⟩"),
        
        # Complex superposition with phase
        (np.pi/2, np.pi/2, "|+Y⟩ Complex Superposition", "50% |0⟩, 50% |1⟩"),
        (np.pi/2, 3*np.pi/2, "|-Y⟩ Complex Superposition", "50% |0⟩, 50% |1⟩"),
        
        # Unequal superposition states
        (np.pi/3, 0, "Unequal Superposition 1", "75% |0⟩, 25% |1⟩"),
        (2*np.pi/3, 0, "Unequal Superposition 2", "25% |0⟩, 75% |1⟩"),
        (np.pi/4, np.pi/4, "General Superposition", "~85% |0⟩, ~15% |1⟩")
    ]
    
    print("\nSuperposition Analysis:")
    print("State\t\t\tθ\t\tφ\t\tAmplitudes\t\t\tProbabilities\t\tMeasured")
    print("-" * 100)
    
    for theta, phi, description, expected in superposition_tests:
        # Calculate theoretical amplitudes from transcript formulas
        alpha_theoretical = np.cos(theta/2)                    # cos(θ/2) for |0⟩
        beta_theoretical = np.exp(1j * phi) * np.sin(theta/2)  # e^(iφ)sin(θ/2) for |1⟩
        
        # Calculate theoretical probabilities (squared magnitudes)
        p0_theoretical = abs(alpha_theoretical)**2  # |cos(θ/2)|²
        p1_theoretical = abs(beta_theoretical)**2   # |sin(θ/2)|²
        
        # Create quantum circuit to verify
        circuit = Circuit()
        if theta == 0 and phi == 0:
            circuit.i(0)  # Identity for |0⟩ state
        else:
            circuit.ry(0, theta)  # Y-rotation for theta
            if abs(phi) > 1e-10:
                circuit.rz(0, phi)  # Z-rotation for phi
        circuit.state_vector()
        
        # Get quantum results
        result = device.run(circuit, shots=0).result()
        statevector = np.array(result.values[0])
        alpha_quantum, beta_quantum = statevector[0], statevector[1]
        
        # Calculate quantum probabilities
        p0_quantum = abs(alpha_quantum)**2
        p1_quantum = abs(beta_quantum)**2
        
        # Verify measurement with sampling
        measurement_circuit = Circuit()
        if theta == 0 and phi == 0:
            measurement_circuit.i(0)
        else:
            measurement_circuit.ry(0, theta)
            if abs(phi) > 1e-10:
                measurement_circuit.rz(0, phi)
        
        measurement_result = device.run(measurement_circuit, shots=10000).result()
        counts = measurement_result.measurement_counts
        total_shots = sum(counts.values())
        p0_measured = counts.get('0', 0) / total_shots
        p1_measured = counts.get('1', 0) / total_shots
        
        print(f"{description[:20]:<20}\t{theta:.3f}\t\t{phi:.3f}\t\t[{alpha_quantum:.3f}, {beta_quantum:.3f}]\t\t({p0_quantum:.3f}, {p1_quantum:.3f})\t\t({p0_measured:.3f}, {p1_measured:.3f})")
        
        # Verify transcript formulas
        p0_formula = (1 + np.cos(theta))/2  # From transcript: (1 + cos θ)/2
        p1_formula = (1 - np.cos(theta))/2  # From transcript: (1 - cos θ)/2
        
        formula_error = abs(p0_quantum - p0_formula) + abs(p1_quantum - p1_formula)
        if formula_error > 1e-10:
            print(f"  ⚠ Formula verification error: {formula_error:.2e}")

def demonstrate_equal_superposition_states():
    """
    Demonstrate the equal superposition states mentioned in the transcript
    """
    
    print("\n=== EQUAL SUPERPOSITION STATES FROM TRANSCRIPT ===")
    print("From transcript: 'plus minus X...plus minus Y...equal superposition of 0 and one'")
    
    # Create the four equal superposition states from transcript
    equal_superposition_states = {
        '|+⟩ (X eigenstate +1)': {
            'circuit': Circuit().h(0).state_vector(),
            'theoretical': np.array([1, 1]) / np.sqrt(2),
            'description': 'H|0⟩ = (|0⟩ + |1⟩)/√2'
        },
        '|-⟩ (X eigenstate -1)': {
            'circuit': Circuit().x(0).h(0).state_vector(),
            'theoretical': np.array([1, -1]) / np.sqrt(2),
            'description': 'H|1⟩ = (|0⟩ - |1⟩)/√2'
        },
        '|+Y⟩ (Y eigenstate +1)': {
            'circuit': Circuit().h(0).s(0).state_vector(),
            'theoretical': np.array([1, 1j]) / np.sqrt(2),
            'description': '(|0⟩ + i|1⟩)/√2'
        },
        '|-Y⟩ (Y eigenstate -1)': {
            'circuit': Circuit().h(0).s(0).z(0).state_vector(),
            'theoretical': np.array([1, -1j]) / np.sqrt(2),
            'description': '(|0⟩ - i|1⟩)/√2'
        }
    }
    
    print("State\t\t\tAmplitudes\t\t\tProbabilities\t\tPhase Relationship")
    print("-" * 85)
    
    for state_name, state_info in equal_superposition_states.items():
        # Get quantum state
        result = device.run(state_info['circuit'], shots=0).result()
        quantum_state = np.array(result.values[0])
        theoretical_state = state_info['theoretical']
        
        # Calculate probabilities
        p0 = abs(quantum_state[0])**2
        p1 = abs(quantum_state[1])**2
        
        # Analyze phase relationship
        alpha, beta = quantum_state[0], quantum_state[1]
        if abs(alpha) > 1e-10 and abs(beta) > 1e-10:
            relative_phase = np.angle(beta) - np.angle(alpha)
            if relative_phase < 0:
                relative_phase += 2*np.pi
        else:
            relative_phase = 0
        
        # Determine phase description
        if abs(relative_phase) < 0.1:
            phase_desc = "In phase (0°)"
        elif abs(relative_phase - np.pi) < 0.1:
            phase_desc = "Out of phase (180°)"
        elif abs(relative_phase - np.pi/2) < 0.1:
            phase_desc = "Quadrature (+90°)"
        elif abs(relative_phase - 3*np.pi/2) < 0.1:
            phase_desc = "Quadrature (-90°)"
        else:
            phase_desc = f"Phase: {np.degrees(relative_phase):.1f}°"
        
        print(f"{state_name[:20]:<20}\t[{quantum_state[0]:.3f}, {quantum_state[1]:.3f}]\t\t({p0:.3f}, {p1:.3f})\t\t{phase_desc}")
        
        # Verify equal superposition property
        if abs(p0 - 0.5) > 0.01 or abs(p1 - 0.5) > 0.01:
            print(f"  ⚠ Not equal superposition: P(0)={p0:.3f}, P(1)={p1:.3f}")

def analyze_measurement_basis_effects():
    """
    Analyze how measurement basis affects observed probabilities
    """
    
    print("\n=== MEASUREMENT BASIS EFFECTS ===")
    print("Demonstrating how the same quantum state gives different results in different measurement bases")
    
    # Create a test state: |ψ⟩ = |+⟩ = (|0⟩ + |1⟩)/√2
    test_state_circuit = Circuit().h(0).state_vector()
    result = device.run(test_state_circuit, shots=0).result()
    test_statevector = np.array(result.values[0])
    
    print(f"Test state |+⟩: [{test_statevector[0]:.4f}, {test_statevector[1]:.4f}]")
    
    # Measurement in different bases
    measurement_bases = {
        'Computational (Z) basis': {
            'circuit': Circuit().h(0),  # Prepare |+⟩ and measure directly
            'description': 'Measure in {|0⟩, |1⟩} basis'
        },
        'X basis': {
            'circuit': Circuit().h(0).h(0),  # Prepare |+⟩, rotate to X basis, measure
            'description': 'Measure in {|+⟩, |-⟩} basis'
        },
        'Y basis': {
            'circuit': Circuit().h(0).s(0).h(0),  # Prepare |+⟩, rotate to Y basis, measure
            'description': 'Measure in {|+Y⟩, |-Y⟩} basis'
        }
    }
    
    print("\nMeasurement Basis Analysis:")
    print("Basis\t\t\tMeasurement Results\t\tExpected\t\tExplanation")
    print("-" * 90)
    
    for basis_name, basis_info in measurement_bases.items():
        # Perform measurement
        measurement_result = device.run(basis_info['circuit'], shots=10000).result()
        counts = measurement_result.measurement_counts
        total_shots = sum(counts.values())
        
        p0_measured = counts.get('0', 0) / total_shots
        p1_measured = counts.get('1', 0) / total_shots
        
        # Provide expected results and explanations
        if 'Computational' in basis_name:
            expected = "50% each"
            explanation = "|+⟩ is superposition in Z basis"
        elif 'X basis' in basis_name:
            expected = "100% |0⟩"
            explanation = "|+⟩ is eigenstate of X"
        elif 'Y basis' in basis_name:
            expected = "50% each"
            explanation = "|+⟩ is superposition in Y basis"
        
        print(f"{basis_name[:20]:<20}\t({p0_measured:.3f}, {p1_measured:.3f})\t\t{expected:<12}\t{explanation}")

def demonstrate_phase_independence_in_probabilities():
    """
    Demonstrate that measurement probabilities are independent of global phase
    but relative phases matter for interference
    """
    
    print("\n=== PHASE EFFECTS ON MEASUREMENT PROBABILITIES ===")
    print("From transcript: Phase φ affects quantum state but not measurement probabilities")
    
    # Test states with same θ but different φ
    theta_fixed = np.pi/3  # Fixed polar angle
    phi_values = [0, np.pi/4, np.pi/2, np.pi, 3*np.pi/2, 7*np.pi/4]
    
    print(f"Fixed θ = {theta_fixed:.3f}, varying φ:")
    print("φ\t\tState Vector\t\t\t\tProbabilities\t\tBloch (x,y,z)")
    print("-" * 80)
    
    for phi in phi_values:
        # Create state with fixed θ, varying φ
        circuit = Circuit()
        circuit.ry(0, theta_fixed)
        if abs(phi) > 1e-10:
            circuit.rz(0, phi)
        circuit.state_vector()
        
        # Get statevector
        result = device.run(circuit, shots=0).result()
        statevector = np.array(result.values[0])
        
        # Calculate probabilities
        p0 = abs(statevector[0])**2
        p1 = abs(statevector[1])**2
        
        # Calculate Bloch coordinates
        alpha, beta = statevector[0], statevector[1]
        r_x = 2 * np.real(np.conj(alpha) * beta)
        r_y = 2 * np.imag(np.conj(alpha) * beta)
        r_z = abs(alpha)**2 - abs(beta)**2
        
        print(f"{phi:.3f}\t\t[{statevector[0]:.3f}, {statevector[1]:.3f}]\t\t\t({p0:.3f}, {p1:.3f})\t\t({r_x:.3f}, {r_y:.3f}, {r_z:.3f})")
    
    print("\nObservation: Probabilities remain constant, but Bloch coordinates change with φ")

def create_superposition_probability_visualization():
    """
    Create visualization showing relationship between superposition angle and probabilities
    """
    
    print("\n=== CREATING SUPERPOSITION PROBABILITY VISUALIZATION ===")
    
    # Generate data for visualization
    theta_range = np.linspace(0, np.pi, 50)
    probabilities_0 = []
    probabilities_1 = []
    
    for theta in theta_range:
        p0 = (1 + np.cos(theta))/2  # From transcript formula
        p1 = (1 - np.cos(theta))/2  # From transcript formula
        probabilities_0.append(p0)
        probabilities_1.append(p1)
    
    # Create the plot
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))
    
    # Plot 1: Probability vs theta
    ax1.plot(theta_range, probabilities_0, 'b-', linewidth=2, label='P(|0⟩) = (1 + cos θ)/2')
    ax1.plot(theta_range, probabilities_1, 'r-', linewidth=2, label='P(|1⟩) = (1 - cos θ)/2')
    ax1.axhline(y=0.5, color='gray', linestyle='--', alpha=0.7, label='Equal superposition')
    
    # Mark special points from transcript
    special_points = [
        (0, 1, 0, '|0⟩'),
        (np.pi/2, 0.5, 0.5, '|+⟩, |-⟩'),
        (np.pi, 0, 1, '|1⟩')
    ]
    
    for theta_sp, p0_sp, p1_sp, label in special_points:
        ax1.plot(theta_sp, p0_sp, 'bo', markersize=8)
        ax1.plot(theta_sp, p1_sp, 'ro', markersize=8)
        ax1.annotate(label, (theta_sp, max(p0_sp, p1_sp) + 0.05), 
                    ha='center', fontweight='bold')
    
    ax1.set_xlabel('θ (Polar Angle)', fontsize=12)
    ax1.set_ylabel('Measurement Probability', fontsize=12)
    ax1.set_title('Measurement Probabilities vs Superposition Angle', fontsize=14)
    ax1.legend()
    ax1.grid(True, alpha=0.3)
    ax1.set_xlim(0, np.pi)
    ax1.set_ylim(0, 1)
    
    # Plot 2: Bloch sphere cross-section
    ax2.set_aspect('equal')
    circle = plt.Circle((0, 0), 1, fill=False, color='gray', alpha=0.5)
    ax2.add_patch(circle)
    
    # Plot states at different theta values
    theta_samples = [0, np.pi/4, np.pi/2, 3*np.pi/4, np.pi]
    colors = ['blue', 'green', 'orange', 'purple', 'red']
    
    for i, theta in enumerate(theta_samples):
        z = np.cos(theta)
        x = np.sin(theta)  # φ = 0 for simplicity
        
        ax2.plot([0, x], [0, z], color=colors[i], linewidth=2, alpha=0.7)
        ax2.plot(x, z, 'o', color=colors[i], markersize=8)
        ax2.annotate(f'θ={theta:.2f}', (x, z), xytext=(5, 5), 
                    textcoords='offset points', fontsize=10)
    
    ax2.set_xlim(-1.2, 1.2)
    ax2.set_ylim(-1.2, 1.2)
    ax2.set_xlabel('X (sin θ)', fontsize=12)
    ax2.set_ylabel('Z (cos θ)', fontsize=12)
    ax2.set_title('Bloch Sphere Cross-Section (X-Z plane)', fontsize=14)
    ax2.grid(True, alpha=0.3)
    
    # Add axis labels
    ax2.text(0, 1.1, '|0⟩', ha='center', fontweight='bold', fontsize=12)
    ax2.text(0, -1.1, '|1⟩', ha='center', fontweight='bold', fontsize=12)
    ax2.text(1.1, 0, '|+⟩', ha='center', fontweight='bold', fontsize=12)
    
    plt.tight_layout()
    plt.show()

# Execute the complete analysis
def main():
    """Execute complete superposition and probability analysis"""
    
    print("=== CONCEPT 5: QUANTUM SUPERPOSITION AND MEASUREMENT PROBABILITIES ===")
    print("Based on SRT transcript analysis of quantum measurement relationships\n")
    
    # Analyze superposition probabilities
    analyze_superposition_probabilities()
    
    # Demonstrate equal superposition states
    demonstrate_equal_superposition_states()
    
    # Analyze measurement basis effects
    analyze_measurement_basis_effects()
    
    # Demonstrate phase independence
    demonstrate_phase_independence_in_probabilities()
    
    # Create visualization
    create_superposition_probability_visualization()
    
    print("\n=== KEY INSIGHTS FROM CONCEPT 5 ===")
    print("✓ Measurement probabilities = squared magnitudes of amplitudes")
    print("✓ P(|0⟩) = |cos(θ/2)|² = (1 + cos θ)/2")
    print("✓ P(|1⟩) = |sin(θ/2)|² = (1 - cos θ)/2")
    print("✓ Equal superposition states have 50% probability for each outcome")
    print("✓ Phase φ affects quantum state but not measurement probabilities")
    print("✓ Measurement basis determines which superposition is observable")
    print("✓ Quantum interference depends on relative phases between amplitudes")

# Run the analysis
main()
```
